```python
from tetra.utils import viewer
import os
import matplotlib.pyplot as plt
import random
```

```python
import tetra.utils.viewer as viewer
print(viewer.__file__)
```

```python
# Path
image_dir = "../dataset_sliced/split_1/train/images"
label_dir = "../dataset_sliced/split_1/train/labels"
class_file = "../classes.txt"
```

```python
#Load files
all_files = [f for f in os.listdir(image_dir) if f.endswith("png") or f.endswith("jpg")]
```

```python
#Display images
#Define the number of images to display
n = 50
#indexes = [random.randint(0,len(all_files) - 1 ) for i in range(n)]
indexes = (50,52)
for f in indexes:
    f = all_files[f]
    image_path = os.path.join(image_dir, f)
    label_path = os.path.join(label_dir, os.path.splitext(f)[0] + ".txt")
    image = viewer.draw_boxes(image_path, label_path,class_file)
    plt.imshow(image)
    plt.title(f)
    plt.axis("off")
    plt.show()
```
