```python
from PIL import Image
import os
from tetra.model import inference
from tetra.utils import file_io, viewer
import sahi
import matplotlib.pyplot as plt
from pathlib import Path
import importlib
```

```python
model_path = os.path.abspath("../../runs/train/lyrurus_yolov11l/weights/best.pt")
image_path = os.path.abspath("../../dataset")  
save_dir = os.path.abspath("../../outputs/labels")
class_file = os.path.abspath("../classes.txt")
```

```python
importlib.reload(file_io)
```

```python
#Get the images
images = list(Path(image_path).glob("*.jpg"))[:10]

for img in images:
    if img.suffix.lower() == ".jpg":
        res = inference.inference(model_path=model_path, image_path=str(img))
        # Convert to YOLO
        yolo_labels = file_io.coco_to_yolo(res.to_coco_annotations(), str(img))
        
        # Get image name (without extension)
        image_name = img.stem + ".txt"
        
        # Save labels
        file_io.save_yolo_labels(yolo_labels, save_dir, image_name)
```

```python
for img in Path(image_path).iterdir():
    if img.suffix.lower() == ".jpg":
        print(img.stem + "text")
        image = viewer.view_image(img, os.path.join(save_dir,img.stem + ".txt"), class_file,0.385)
        plt.imshow(image)
        plt.title("Inference")
        plt.axis("off")
        plt.show()
```
